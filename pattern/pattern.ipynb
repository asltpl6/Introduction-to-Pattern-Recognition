```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
import sklearn
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=27)
```

```python
print(X_train)  
print(y_train)
```

```python
KNN_model = KNeighborsClassifier(n_neighbors=5)
```

```python
 KNN_model.fit(X_train, y_train)
```

```python
KNN_prediction = KNN_model.predict(X_test)
```

```python
print(accuracy_score(KNN_prediction,y_test)
```

```python
#Confusion matrix, Accuracy, sensitivity and specificity
conf_matrix = confusion_matrix(y_true=y_test, y_pred=KNN_prediction)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.matshow(conf_matrix, cmap=plt.cm.Oranges, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')
 
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

```python

total1=sum(sum(conf_matrix))
#####from confusion matrix calculate accuracy
accuracy1=(conf_matrix[0,0]+conf_matrix[1,1])/total1
print ('Accuracy : ', accuracy1)

sensitivity1 = conf_matrix[0,0]/(conf_matrix[0,0]+conf_matrix[0,1])
print('Sensitivity : ', sensitivity1 )

specificity1 = conf_matrix[1,1]/(conf_matrix[1,0]+conf_matrix[1,1])
print('Specificity : ', specificity1)
auc = roc_auc_score(y_test, KNN_model.predict_proba(X_test)[:, 1])
print('AUC : ', auc)
```
